```python
import pandas as pd

data = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
data
```

```python
from sklearn.model_selection import train_test_split


x2=data.drop('Class',axis=1)
y2=data['Class']
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns_to_normalize = [f'V{i}' for i in range(1, 29)]

scaler.fit(data[columns_to_normalize])

data[columns_to_normalize] = scaler.transform(data[columns_to_normalize])

data
```

```python
class_counts = data['Class'].value_counts()
class_counts
```

```python
correlation_matrix = data.corr()
price_correlation = correlation_matrix['Class'].sort_values(ascending=False)
price_correlation
```

```python
from operator import xor


# Separate the data by class
class_0 = data[data['Class'] == 0]
class_1 = data[data['Class'] == 1]

# Downsample the majority class
class_0_downsampled = class_0.sample(n=5000, random_state=42)

# Concatenate the downsampled majority class with the minority class
balanced_data = pd.concat([class_0_downsampled, class_1])

# Shuffle the balanced dataset
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)


x=balanced_data.drop('Class',axis=1)
y=balanced_data['Class']
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn_model = KNeighborsClassifier(n_neighbors=5)  
knn_model.fit(X_train, y_train)

knn_y_pred = knn_model.predict(X_test)
knn_y2_pred = knn_model.predict(x2_test)

knn_accuracy = accuracy_score(y_test, knn_y_pred)
print(f"KNN Accuracy: {knn_accuracy}")
print(classification_report(y_test, knn_y_pred))

knn_accuracy2 = accuracy_score(y2_test, knn_y2_pred)
print(f"KNN Accuracy_2: {knn_accuracy2}")
print(classification_report(y2_test, knn_y2_pred))


for i in range(10): 
  print(f"Sample {i+1}: Actual Class - {y_test.iloc[i]}, Predicted Class - {knn_y2_pred[i]}")
```
